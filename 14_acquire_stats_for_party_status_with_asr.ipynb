```python
from utils import example_eval_config, example_train_config, eval_model

results = []
clip = -1
OUTPUT_DIR = "models/classla_wav2vec2-large-slavic-parlaspeech-hr_PARTY_STATUS_" 
if clip == 2:
    OUTPUT_DIR += "2s_"

for split in "dev test".split():
    config = {
        "output_column": "Party_status",
        "model_name_or_path": OUTPUT_DIR+"/checkpoint-4680",
        "eval_file": f"012_{split}.csv",
        "clip_seconds": clip,
    }
    y_true, y_pred = eval_model(config)

    results.append({**config, "y_true": y_true, "y_pred": y_pred})
```

```python
import pandas as pd
df = pd.DataFrame(data=results)


from sklearn.metrics import accuracy_score, f1_score

df["accuracy"] = df.apply(
    lambda row: accuracy_score(row["y_true"], row["y_pred"]), axis=1
)
df["macroF1"] = df.apply(
    lambda row: f1_score(row["y_true"], row["y_pred"], average="macro"), axis=1
)
df
```

```python
print(df.to_json(orient="records", lines=True))
```
